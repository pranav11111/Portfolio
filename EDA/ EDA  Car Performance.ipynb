```python
fig = plt.figure(figsize=(6,6))
plt.pie(car["origin"].value_counts(),labels =car["origin"].value_counts().index, autopct="%0.1f%%")
```

```python
plt.subplots(2,2,figsize=(16,10))

plt.subplot(2,2,1)
plt.pie(x=car["origin"].value_counts(),labels=car["origin"].value_counts().index,autopct='%0.2f%%',data=car)
plt.title('Percentage of each origin in the data')

plt.subplot(2,2,2)
plt.title("Distribution of cylinder")
plt.pie(x=car["cylinders"].value_counts(),labels=car["cylinders"].value_counts().index,autopct='%0.2f%%',explode=[0,0,0,0.03,0.2],data=car)


plt.subplot(2,2,3)

plt.title("Distribution of model_year")
plt.pie(x=car["model_year"].value_counts(),labels=car["model_year"].value_counts().index,autopct='%0.2f%%',data=car)

plt.subplot(2,2,4)
plt.title("Distribution of mpg_level")
plt.pie(x=car["mpg_level"].value_counts(),labels=car["mpg_level"].value_counts().index,autopct='%0.2f%%',data=car)

plt.show()
```

- USA sells majority of cars
- Most(9.18) are sold in the year 78
- Medium milage cars are most sold
- 4 cylinder car are produced the most while 5 and 3  are produced the least with 8 and 6 cylinder having almost the same sale

```python
plt.figure(figsize = (5,5))
sns.countplot(x = car["cylinders"],hue =car['origin' ])
plt.title("Distribution of cylinders w.r.t to country")
```

- 3 Cylinder cars are only present in Japan
- 5 cylinder cars are only present in Europe
- 8 cylinder cars are only present in USA
- 4 cylinder cars are distibuted evenly in USA and Japan while Europe can still be compared with it having just slightly less
- 6 Cylinder cars do exist in all three regions but USA is till dominant with Japan and Europe having very low amount of them

```python
plt.figure(figsize = (5,5))
sns.countplot(x = car["mpg_level"],hue =car['origin' ])
plt.title("Milage w.r.t to country")
```

- Japan focuses on high milage in comparion to cylinders
- Usa has more number of cylinders but has lower no of high milage cars 


```python
plt.figure(figsize = (5,5))
sns.countplot(x = car["cylinders"],hue =car['mpg_level' ])
```

- The more the no of cylinders the lower the milage(not always)
- 4 cylinder has the maximum use and efficiency(most high milage cars,3 has no high milage cars)

```python
plt.subplots(2,2,figsize = (10,10))


plt.subplot(2,2,1)
sns.lineplot(x='model_year',y='mpg',hue='origin',data = car)
plt.title('milage w.r.t year')

plt.subplot(2,2,2)
sns.lineplot(x='model_year',y='acceleration',hue='origin',data = car)


plt.subplot(2,2,3)
sns.lineplot(x='model_year',y='horsepower',hue='origin',data = car)

plt.subplot(2,2,4)
sns.lineplot(x='model_year',y='weight',hue='origin',data = car)
```

- as the year progresses the mileage increses and horsepower decreses(usa affected majorly)
- weight was also seen to be reduced in usa but in japan the weight was always low
