## Conclusion
- The games having positve rating have lowest average price
- Most of the games having overwhelmingly postive reviews have high prices 
- overwhelmingly negitive games too have high pricing as compared to postive mixed etc.
   - The reason could be that the game was highliy anticipated due to its higher price and could not live up to its expectation in comparison to the games having postive review
   - The game could have been on the same level as a positve review game but due to its higher prices the game got a very bad review
   - But on the other hand  overwhelmingly postive reviews have high prices too, infact the highest prices which shows us that the games having higher prices are mostly very good 
- The pricing, as well as discount has impact on the reviews of the game 
    - The games having higher mean discount have better reviews as compared to games having lower discounts
