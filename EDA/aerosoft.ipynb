Q6 With all the above steps you can answer questions like: What is the probability of a male customer buying a KP781 treadmill?

```python
def age_group(x):
  if x <= 24 :
    return 'Young Adults'
  elif x <= 30 :
    return 'Early Adults'
  elif x <= 40 :
    return 'Adults'
  else :
    return 'Late Adults'

data['AgeGroup'] = data['Age'].apply(age_group)

age_df = pd.crosstab(columns = data['AgeGroup'], index = data['Product'], margins = True, margins_name = 'Total')
age_vals = np.sort(data['AgeGroup'].unique().tolist())
for i in age_vals :
  col_name = str(i) + " probabilty"
  age_df[col_name] = round(age_df[i]/age_df['Total'], 2)
age_df.drop('Total', axis = 0, inplace = True)
age_df
```
